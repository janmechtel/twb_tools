```
#@title
import xml.etree.ElementTree as ET
from google_trans_new import google_translator
from six.moves import urllib
import pandas as pd
import re
import time
import glob
import ntpath
```

```
#@title
UPLOAD_FOLDER = 'gdrive/My Drive/twb_translate/upload'
OUTPUT_FOLDER = 'gdrive/My Drive/twb_translate/output'
```

```
#@title
upload_twbs = glob.glob(UPLOAD_FOLDER + '/*.twb')
output_twbs = glob.glob(OUTPUT_FOLDER + '/*.twb')
```

```
#@title
# translate the text in a twb
def translate_twb(file,src,dest):
    output = OUTPUT_FOLDER + '/' + f'{dest}__' + ntpath.basename(file)  # set the output path
    translator = google_translator()
    # parse the twb file
    tree = ET.parse(file)
    root = tree.getroot()

    # dictionary of translations
    trans_dict = {}
    
    # translate sheet names
    print('***translating sheet names***')
    for child in root.findall('./worksheets/worksheet'):
        # replace some symbols e.g. _ with spaces
        clean_text = child.get('name').replace('_', ' ')
        if child.get('name') in trans_dict:  # check if already translated
          trans_obj = trans_dict[child.get('name')]  # if so, get translation
        else:
          trans_obj = []
          for word in clean_text.split():
            trans_obj.append(translator.translate(word, lang_src=src, lang_tgt=dest))  # translate
            time.sleep(1)
          trans_obj = " ".join(trans_obj)
          print(clean_text, '-->', trans_obj)
          trans_dict[child.get('name')] = trans_obj  # store translation in dict
          time.sleep(1)  # sleep to avoid translation rate limit
        child.attrib['name'] = trans_obj

    for child in root.findall('./dashboards/dashboard/zone'):
        if child.get('name') is not None:
            clean_text = child.get('name').replace('_', ' ')
            if child.get('name') in trans_dict:
              trans_obj = trans_dict[child.get('name')]  # if so, get translation
            else:
              trans_obj = []
              for word in clean_text.split():
                trans_obj.append(translator.translate(word, lang_src=src, lang_tgt=dest))  # translate
                time.sleep(1)
              trans_obj = " ".join(trans_obj)
              print(clean_text, '-->', trans_obj)
              trans_dict[child.get('name')] = trans_obj  # store translation in dict
              time.sleep(1)
            child.attrib['name'] = trans_obj

    for child in root.findall('./dashboards/dashboard/zones'):
        for subchild in child.iter('zone'):
            if subchild.get('name') is not None:
                clean_text = subchild.get('name').replace('_', ' ')
                if subchild.get('name') in trans_dict:
                  trans_obj = trans_dict[subchild.get('name')]
                else:
                  trans_obj = []
                  for word in clean_text.split():
                    trans_obj.append(translator.translate(word, lang_src=src, lang_tgt=dest))  # translate
                    time.sleep(1)
                  trans_obj = " ".join(trans_obj)
                  print(clean_text, '-->', trans_obj)
                  trans_dict[subchild.get('name')] = trans_obj
                  time.sleep(1)
                subchild.attrib['name'] = trans_obj

    for subroot in root.findall('./dashboards/dashboard/devicelayouts/devicelayout/zones'):
        for child in subroot.iter('zone'):
            if child.get('name') is not None:
                clean_text = child.get('name').replace('_', ' ')
                if child.get('name') in trans_dict:
                  trans_obj = trans_dict[child.get('name')]  # if so, get translation
                else:
                  trans_obj = []
                  for word in clean_text.split():
                    trans_obj.append(translator.translate(word, lang_src=src, lang_tgt=dest))  # translate
                    time.sleep(1)
                  trans_obj = " ".join(trans_obj)
                  print(clean_text, '-->', trans_obj)
                  trans_dict[child.get('name')] = trans_obj  # store translation in dict
                  time.sleep(1)
                child.attrib['name'] = trans_obj

    for child in root.iter('viewpoint'):
        if child.get('name') is not None:
            clean_text = child.get('name').replace('_', ' ')
            if child.get('name') in trans_dict:
              trans_obj = trans_dict[child.get('name')]  # if so, get translation
            else:
              trans_obj = []
              for word in clean_text.split():
                trans_obj.append(translator.translate(word, lang_src=src, lang_tgt=dest))  # translate
                time.sleep(1)
              trans_obj = " ".join(trans_obj)
              print(clean_text, '-->', trans_obj)
              trans_dict[child.get('name')] = trans_obj  # store translation in dict
              time.sleep(1)
            child.attrib['name'] = trans_obj

    for child in root.iter('story-point'):
        if child.get('captured-sheet') is not None:
            clean_text = child.get('captured-sheet').replace('_', ' ')
            if child.get('captured-sheet') in trans_dict:
              trans_obj = trans_dict[child.get('captured-sheet')]  # if so, get translation
            else:
              trans_obj = []
              for word in clean_text.split():
                trans_obj.append(translator.translate(word, lang_src=src, lang_tgt=dest))  # translate
                time.sleep(1)
              trans_obj = " ".join(trans_obj)
              print(clean_text, '-->', trans_obj)
              trans_dict[child.get('captured-sheet')] = trans_obj  # store translation in dict
              time.sleep(1)
            child.attrib['captured-sheet'] = trans_obj

    # translate dashboard and story names
    print('***translating dashboard names***')
    for child in root.findall('./dashboards/dashboard'):
        clean_text = child.get('name').replace('_', ' ')
        if child.get('name') in trans_dict:
          trans_obj = trans_dict[child.get('name')]  # if so, get translation
        else:
          trans_obj = []
          for word in clean_text.split():
            trans_obj.append(translator.translate(word, lang_src=src, lang_tgt=dest))  # translate
            time.sleep(1)
          trans_obj = " ".join(trans_obj)
          print(clean_text, '-->', trans_obj)
          trans_dict[child.get('name')] = trans_obj  # store translation in dict
          time.sleep(1)
        child.attrib['name'] = trans_obj

    # translate story point titles
    print('***translating story point titles***')
    for child in root.iter('story-point'):
        clean_text = child.get('caption').replace('_', ' ')
        if child.get('caption') in trans_dict:
          trans_obj = trans_dict[child.get('caption')]  # if so, get translation
        else:
          trans_obj = []
          for word in clean_text.split():
            trans_obj.append(translator.translate(word, lang_src=src, lang_tgt=dest))  # translate
            time.sleep(1)
          trans_obj = " ".join(trans_obj)
          print(clean_text, '-->', trans_obj)
          trans_dict[child.get('caption')] = trans_obj  # store translation in dict
          time.sleep(1)
        child.attrib['caption'] = trans_obj

    # translate all text objects, including dashboard and story point text boxes,
    # sheet captions, and custom filter titles on dashboards
    print('***translating text objects***')
    for child in root.iter('run'):
        if not (child.text.startswith('<') or child.text.startswith('&')
                or child.text.startswith('Æ') or '.[Multiple Values]' in child.text
                or child.text.startswith('[') or '[federated' in child.text):
            # replace some symbols e.g. _ with spaces
            clean_text = child.text.replace('_', ' ')
            if child.text in trans_dict:
              trans_obj = trans_dict[child.text]
            else:
              trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
              trans_dict[child.text] = trans_obj
              print(clean_text, '-->', trans_obj)
              time.sleep(1)
            child.text = trans_obj

    print('***translating text objects 1***')
    for child in root.iter('format'):
        if child.get('attr') == 'title' and len(child.get('value')) > 1:
            #replace some symbols e.g. _ with spaces
            clean_text = child.get('value').replace('_', ' ')
            if child.get('value') in trans_dict:
              trans_obj = trans_dict[child.get('value')]  # if so, get translation
            else:
              trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
              print(clean_text, '-->', trans_obj)
              trans_dict[child.get('value')] = trans_obj  # store translation in dict
              time.sleep(1)
            child.attrib['value'] = trans_obj

    print('***translating text objects 2***')
    for child in root.findall('./datasources/datasource/column'):
        if child.get('alias') is not None:
            clean_text = child.get('alias').replace('_', ' ')
            if child.get('alias') in trans_dict:
              trans_obj = trans_dict[child.get('alias')]  # if so, get translation
            else:
              trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
              print(clean_text, '-->', trans_obj)
              trans_dict[child.get('alias')] = trans_obj  # store translation in dict
              time.sleep(1)
            child.attrib['alias'] = trans_obj

    print('***translating text objects 3***')
    for child in root.findall('./datasources/datasource/column/aliases/alias'):
        if child.get('value') is not None:
            clean_text = child.get('value').replace('_', ' ')
            if child.get('value') in trans_dict:
              trans_obj = trans_dict[child.get('value')]  # if so, get translation
            else:
              trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
              print(clean_text, '-->', trans_obj)
              trans_dict[child.get('value')] = trans_obj  # store translation in dict
              time.sleep(1)
            child.attrib['value'] = trans_obj

    print('***translating text objects 4***')
    for child in root.findall('./datasources/datasource/column/members/member'):
        if child.get('alias') is not None:
            clean_text = child.get('alias').replace('_', ' ')
            if child.get('alias') in trans_dict:
              trans_obj = trans_dict[child.get('alias')]  # if so, get translation
            else:
              trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
              print(clean_text, '-->', trans_obj)
              trans_dict[child.get('alias')] = trans_obj  # store translation in dict
              time.sleep(1)
            child.attrib['alias'] = trans_obj

    print('***translating text objects 5***')
    for child in root.findall('./datasources/datasource/datasource-dependencies/column'):
        if child.get('alias') is not None:
            clean_text = child.get('alias').replace('_', ' ')
            if child.get('alias') in trans_dict:
              trans_obj = trans_dict[child.get('alias')]  # if so, get translation
            else:
              trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
              print(clean_text, '-->', trans_obj)
              trans_dict[child.get('alias')] = trans_obj  # store translation in dict
              time.sleep(1)
            child.attrib['alias'] = trans_obj

    print('***translating text objects 6***')
    for child in root.findall('./datasources/datasource/datasource-dependencies/column/aliases/alias'):
        if child.get('value') is not None:
            clean_text = child.get('value').replace('_', ' ')
            if child.get('value') in trans_dict:
              trans_obj = trans_dict[child.get('value')]  # if so, get translation
            else:
              trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
              print(clean_text, '-->', trans_obj)
              trans_dict[child.get('value')] = trans_obj  # store translation in dict
              time.sleep(1)
            child.attrib['value'] = trans_obj

    print('***translating text objects 7***')
    for child in root.findall('./datasources/datasource/datasource-dependencies/members/member'):
        if child.get('alias') is not None:
            clean_text = child.get('alias').replace('_', ' ')
            if child.get('alias') in trans_dict:
              trans_obj = trans_dict[child.get('alias')]  # if so, get translation
            else:
              trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
              print(clean_text, '-->', trans_obj)
              trans_dict[child.get('alias')] = trans_obj  # store translation in dict
              time.sleep(1)
            child.attrib['alias'] = trans_obj

    print('***translating text objects 8***')
    for child in root.findall('./worksheets/worksheet/table/view/datasource-dependencies/column'):
        if child.get('caption') is not None:
            if child.get('caption') in trans_dict:
                child.attrib['caption'] = trans_dict[child.get('caption')]
            else:
                clean_text = child.get('caption').replace('_', ' ')
                trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
                print(clean_text, '-->', trans_obj)
                time.sleep(1)
                new_text = trans_obj
                trans_dict[child.get('caption')] = new_text
                child.attrib['caption'] = new_text
        else:
            if child.get('name') is not None:
                clean_text = child.get('name').replace('_', ' ')
                if child.get('alias') in trans_dict:
                  trans_obj = trans_dict[child.get('alias')]  # if so, get translation
                else:
                  trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
                  print(clean_text, '-->', trans_obj)
                  trans_dict[child.get('alias')] = trans_obj  # store translation in dict
                  time.sleep(1)
                child.attrib['alias'] = trans_obj

    # # create a list of fields in each ds
    # for parent in root.findall('./datasources/datasource'):
    #     view_cols = []
    #     datasource = parent.get('name')
    #     for child in parent.iter('local-name'):
    #         view_cols.append(child.text)
    #     ds_view_cols[datasource] = view_cols

    # for ds, cols in ds_ds_cols.items():
    #     for key in cols:
    #         print(ds + '- ', key + ': ', cols[key])

    print('***translating text objects 9***')
    for child in root.iter('column'):
        if child.get('caption') is not None:
            if child.get('caption') in trans_dict:
                child.attrib['caption'] = trans_dict[child.get('caption')]
            else:
                clean_text = child.get('caption').replace('_', ' ')
                trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
                print(clean_text, '-->', trans_obj)
                time.sleep(1)
                new_text = trans_obj
                trans_dict[child.get('caption')] = new_text
                child.attrib['caption'] = new_text
        else:
            if child.get('name') is not None:
                clean_text = child.get('name').replace('_', ' ')
                clean_text = clean_text.replace('[', '')
                clean_text = clean_text.replace(']', '')
                trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
                print(clean_text, '-->', trans_obj)
                time.sleep(1)
                new_text = trans_obj
                trans_dict[child.get('name')] = new_text
                child.attrib['caption'] = new_text

    # translate groups and bins
    print('***translating groups and bins***')
    for child in root.findall('./datasources/datasource/column'):
        for tag in child.findall('calculation[@class]'):
            if tag.get('class') == 'categorical-bin' or tag.get('class') == 'bin':
                if child.get('caption') is not None:
                    clean_text = child.get('caption').replace('_', ' ')
                    if child.get('caption') in trans_dict:
                      trans_obj = trans_dict[child.get('caption')]  # if so, get translation
                    else:
                      trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
                      print(clean_text, '-->', trans_obj)
                      trans_dict[child.get('caption')] = trans_obj  # store translation in dict
                      time.sleep(1)
                    child.attrib['caption'] = trans_obj

    # translate sets
    print('***translating sets***')
    for child in root.findall('./datasources/datasource/group'):
        if child.get('caption') is not None:
            clean_text = child.get('caption').replace('_', ' ')
            if child.get('caption') in trans_dict:
              trans_obj = trans_dict[child.get('caption')]  # if so, get translation
            else:
              trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
              print(clean_text, '-->', trans_obj)
              trans_dict[child.get('caption')] = trans_obj  # store translation in dict
              time.sleep(1)
            child.attrib['caption'] = trans_obj

    for child in root.findall('./datasources/datasource/group/groupfilter/groupfilter'):
        if child.get('function') == 'reference':
            if child.get('field') is not None:
                clean_text = child.get('field').replace('_', ' ')
                clean_text = clean_text.replace('[', '')
                clean_text = clean_text.replace(']', '')
                if child.get('field') in trans_dict:
                  trans_obj = trans_dict[child.get('field')]
                else:
                  trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
                  time.sleep(1)
                  trans_dict[child.get('field')] = '[' + trans_obj + ']'
                child.attrib['field'] = '[' + trans_obj + ']'

    # translate folders
    print('***translating folders***')
    for child in root.findall('./datasources/datasource/*folder'):
        clean_text = child.get('name').replace('_', ' ')
        if child.get('name') in trans_dict:
          trans_obj = trans_dict[child.get('name')]  # if so, get translation
        else:
          trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
          print(clean_text, '-->', trans_obj)
          trans_dict[child.get('name')] = trans_obj  # store translation in dict
          time.sleep(1)
        child.attrib['name'] = trans_obj

    for child in root.findall('./datasources/datasource/*/folder'):
        clean_text = child.get('name').replace('_', ' ')
        if child.get('name') in trans_dict:
          trans_obj = trans_dict[child.get('name')]  # if so, get translation
        else:
          trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
          print(clean_text, '-->', trans_obj)
          trans_dict[child.get('name')] = trans_obj  # store translation in dict
          time.sleep(1)
        child.attrib['name'] = trans_obj

    # update parameter dependencies
    print('***translating parameter dependencies***')
    for child in root.findall('./datasources/datasource/datasource-dependencies/column'):
        clean_text = child.get('caption').replace('_', ' ')
        if child.get('caption') in trans_dict:
          trans_obj = trans_dict[child.get('caption')]  # if so, get translation
        else:
          trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
          print(clean_text, '-->', trans_obj)
          trans_dict[child.get('caption')] = trans_obj  # store translation in dict
          time.sleep(1)
        child.attrib['caption'] = trans_obj

    # # update windows
    # print('***translating windows***')
    # for child in root.findall('./windows/window'):
    #     clean_text = child.get('name').replace('_', ' ')
    #     if child.get('name') in trans_dict:
    #       trans_obj = trans_dict[child.get('name')]  # if so, get translation
    #     else:
    #       trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
    #       print(clean_text, '-->', trans_obj)
    #       trans_dict[child.get('name')] = trans_obj  # store translation in dict
    #       time.sleep(1)
    #     child.attrib['name'] = trans_obj

    # update thumbnails
    print('***translating thumbnails***')
    for child in root.findall('./thumbnails/thumbnail'):
        clean_text = child.get('name').replace('_', ' ')
        if child.get('name') in trans_dict:
          trans_obj = trans_dict[child.get('name')]  # if so, get translation
        else:
          trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
          print(clean_text, '-->', trans_obj)
          trans_dict[child.get('name')] = trans_obj  # store translation in dict
          time.sleep(1)
        child.attrib['name'] = trans_obj

    # # translate Action filters
    # for child in root.findall('./worksheets/worksheet/table/view/filter'):
    #     if child.get('column') is not None:
    #         s = child.get('column')
    #         pre_action_text = (re.search(r'(?<=\[).*?(?=\])', s).group(0))
    #         action_text = (re.search(r'(?<=\.\[).*?(?=\])', s).group(0))
    #         if action_text.startswith('Action '):
    #             action_field_list = re.search(r'(?<=\().*?(?=\))', s).group(0)
    #             action_field_list = action_field_list.split(',')
    #             translated_list = []
    #             for item in action_field_list:
    #                 if item in trans_dict:
    #                     translated_list.append(trans_dict[item])
    #                 else:
    #                     translated_list.append(item)
    #             new_text = '[' + pre_action_text + '].[' + 'Action (' + ','.join(translated_list) + ')]'
    #             trans_dict[child.get('column')] = new_text
    #             child.attrib['column'] = new_text

    # for child in root.findall('./datasources/datasource/group'):
    #     if child.get('name').startswith('[Action ('):
    #         s = child.get('name')
    #         action_text = (re.search(r'(?<=\[).*?(?=\])', s).group(0))
    #         if action_text.startswith('Action '):
    #             action_field_list = re.search(r'(?<=\().*?(?=\))', s).group(0)
    #             action_field_list = action_field_list.split(',')
    #             translated_list = []
    #             for item in action_field_list:
    #                 if item in trans_dict:
    #                     translated_list.append(trans_dict[item])
    #                 else:
    #                     translated_list.append(item)
    #             new_text = '[' + 'Action (' + ','.join(translated_list) + ')]'
    #             trans_dict[child.get('name')] = new_text
    #             child.attrib['name'] = new_text

    # for child in root.findall('./worksheets/worksheet/table/view/slices/column'):
    #     s = child.text
    #     pre_action_text = (re.search(r'(?<=\[).*?(?=\])', s).group(0))
    #     action_text = (re.search(r'(?<=\.\[).*?(?=\])', s).group(0))
    #     if action_text.startswith('Action '):
    #         action_field_list = re.search(r'(?<=\().*?(?=\))', s).group(0)
    #         action_field_list = action_field_list.split(',')
    #         translated_list = []
    #         for item in action_field_list:
    #             if item in trans_dict:
    #                 translated_list.append(trans_dict[item])
    #             else:
    #                 translated_list.append(item)
    #         new_text = '[' + pre_action_text + '].[' + 'Action (' + ','.join(translated_list) + ')]'
    #         trans_dict[child.text] = new_text
    #         child.text = new_text

    # for child in root.findall('actions/action/source'):
    #     if child.get('dashboard') is not None:
    #         if child.get('dashboard') in trans_dict:
    #             child.attrib['dashboard'] = trans_dict[child.get('dashboard')]
    #         else:
    #             clean_text = child.get('dashboard').replace('_', ' ')
    #             trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
    #             print(clean_text, '-->', trans_obj)
    #             time.sleep(1)
    #             new_text = trans_obj
    #             trans_dict[child.get('dashboard')] = new_text
    #             child.attrib['dashboard'] = new_text
    #     if child.get('worksheet') is not None:
    #         if child.get('worksheet') in trans_dict:
    #             child.attrib['worksheet'] = trans_dict[child.get('worksheet')]
    #         else:
    #             clean_text = child.get('worksheet').replace('_', ' ')
    #             trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
    #             print(clean_text, '-->', trans_obj)
    #             time.sleep(1)
    #             new_text = trans_obj
    #             trans_dict[child.get('worksheet')] = new_text
    #             child.attrib['worksheet'] = new_text

    for child in root.iter('exclude-sheet'):
        if child.get('name') is not None:
            if child.get('name') in trans_dict:
                child.attrib['name'] = trans_dict[child.get('name')]
            else:
                clean_text = child.get('name').replace('_', ' ')
                trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
                print(clean_text, '-->', trans_obj)
                time.sleep(1)
                new_text = trans_obj
                trans_dict[child.get('name')] = new_text
                child.attrib['name'] = new_text

    # for child in root.findall('actions/action/command/param'):
    #     if child.get('name') == 'target':
    #         if child.get('value') in trans_dict:
    #             child.attrib['value'] = trans_dict[child.get('value')]
    #         else:
    #             clean_text = child.get('value').replace('_', ' ')
    #             trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
    #             print(clean_text, '-->', trans_obj)
    #             time.sleep(1)
    #             new_text = trans_obj
    #             trans_dict[child.get('value')] = new_text
    #             child.attrib['value'] = new_text
    #     elif child.get('name') == 'exclude':
    #         if child.get('value') in trans_dict:
    #             child.attrib['value'] = trans_dict[child.get('value')]
    #         else:
    #             clean_text = child.get('value').replace('_', ' ')
    #             trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
    #             print(clean_text, '-->', trans_obj)
    #             time.sleep(1)
    #             new_text = trans_obj
    #             trans_dict[child.get('value')] = new_text
    #             child.attrib['value'] = new_text
    #     elif child.get('name') == 'exclude':
    #         if child.get('value') in trans_dict:
    #             child.attrib['value'] = trans_dict[child.get('value')]
    #         else:
    #             clean_text = child.get('value').replace('_', ' ')
    #             trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
    #             print(clean_text, '-->', trans_obj)
    #             time.sleep(1)
    #             new_text = trans_obj
    #             trans_dict[child.get('value')] = new_text
    #             child.attrib['value'] = new_text

    #     elif child.get('name') == 'exclude':
    #         if child.get('value') in trans_dict:
    #             child.attrib['value'] = trans_dict[child.get('value')]
    #         else:
    #             clean_text = child.get('value').replace('_', ' ')
    #             trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
    #             print(clean_text, '-->', trans_obj)
    #             time.sleep(1)
    #             new_text = trans_obj
    #             trans_dict[child.get('value')] = new_text
    #             child.attrib['value'] = new_text
    #     elif child.get('name') == 'field-captions':
    #         if child.get('value') in trans_dict:
    #             child.attrib['value'] = trans_dict[child.get('value')]
    #         else:
    #             clean_text = child.get('value').replace('_', ' ')
    #             trans_obj = translator.translate(clean_text, lang_src=src, lang_tgt=dest)
    #             print(clean_text, '-->', trans_obj)
    #             time.sleep(1)
    #             new_text = trans_obj
    #             trans_dict[child.get('value')] = new_text
    #             child.attrib['value'] = new_text

    # for child in root.findall('actions/action/link'):
    #     if child.get('expression') is not None:
    #         s = child.get('expression')
    #         snip = re.search(r'(?<=:).*?(?=\?)', s)
    #         if snip is not None:
    #             snip = snip.group(0)
    #             url_decode_snip = urllib.parse.unquote(snip)
    #             if url_decode_snip in trans_dict:
    #                 translated_snip = trans_dict[url_decode_snip]
    #                 url_enc_translated_snip = urllib.parse.quote(translated_snip)
    #                 translated_s = s.replace(snip, url_enc_translated_snip)
    #                 trans_dict[child.get('expression')] = translated_s
    #                 child.attrib['expression'] = translated_s
    #             else:
    #                 trans_obj = translator.translate(url_decode_snip, lang_src=src, lang_tgt=dest)
    #                 print(clean_text, '-->', trans_obj)
    #                 time.sleep(1)
    #                 translated_snip = trans_obj
    #                 url_enc_translated_snip = urllib.parse.quote(translated_snip)
    #                 translated_s = s.replace(snip, url_enc_translated_snip)
    #                 trans_dict[child.get('expression')] = translated_s
    #                 child.attrib['expression'] = translated_s

    with open(output, "wb" ) as f:
      tree.write(f, encoding="UTF-8")
```
